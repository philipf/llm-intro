## Pipeline API

```python
from transformers import pipeline

pipe = pipeline(model="google/gemma-3-1b-pt")
```

Let's use the model!

```python
pipe("The cat sat on the", max_new_tokens=1)
```

```python
output =  pipe("How can I train my puppy to sit?", max_new_tokens=100)
print(output[0]['generated_text'])
```

## Alternative API

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("unsloth/Llama-3.2-1B-bnb-4bit")
model = AutoModelForCausalLM.from_pretrained("unsloth/Llama-3.2-1B-bnb-4bit")
```

```python
prompt = "The cat sat on the"

inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
print("Inputs", inputs)

outputs = model.generate(**inputs, max_new_tokens=1)
print("Outputs", outputs)

decoded_output = tokenizer.decode(outputs[0])
print("Decoded output", decoded_output)
```
